#Theory Questions

Q1 What is a random variable in probability theory?

Ans:- A random variable is a numerical description of the outcome of a statistical experiment. It assigns a real number to each outcome in the sample space.


Q2 What are the types of random variables?

Ans- Discrete Random Variable
- Continuous Random Variable


Q3 What is the difference between discrete and continuous distributions?

- Discrete: Takes specific, separate values (e.g., number of heads in coin toss).
- Continuous: Takes any value in a range (e.g., height, weight).


Q4 What are probability distribution functions (PDF)?

Ans:- A function that provides the probabilities of occurrence of different possible outcomes for a random variable.


Q5 How do cumulative distribution functions (CDF) differ from PDFs?

- PDF shows probability of specific outcomes.
- CDF shows probability that the variable is less than or equal to a value.


Q6 What is a discrete uniform distribution?

Ans- A distribution where each of the n outcomes has equal probability (e.g., rolling a fair die).


Q7 What are the key properties of a Bernoulli distribution?

- Only two outcomes: success (1) or failure (0).
- Probability of success is p; failure is (1-p).


Q8 What is the binomial distribution, and how is it used in probability?

Ans:- Describes the number of successes in a fixed number of independent Bernoulli trials.


Q9 What is the Poisson distribution and where is it applied?

Ans:- Models the number of times an event occurs in a fixed interval of time or space (e.g., number of emails per hour).


Q10 What is a continuous uniform distribution?

Ans:- A distribution where all outcomes in a range are equally likely.


Q11 What are the characteristics of a normal distribution?

- Symmetrical bell-shaped curve.
- Mean = Median = Mode.


Q12 What is the standard normal distribution, and why is it important?

Ans:- A normal distribution with mean 0 and standard deviation 1; it's used for standardizing values and simplifying calculations.


Q13 What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Ans:- States that the sampling distribution of the sample mean approaches a normal distribution as the sample size grows, regardless of the population's distribution.


Q14 How does the Central Limit Theorem relate to the normal distribution?

Ans:- Even if the original data is not normal, the sample mean becomes normally distributed with large samples.


Q15 What is the application of Z statistics in hypothesis testing?

Ans- Used to determine how far (in standard deviations) a data point is from the population mean, to accept or reject hypotheses.


Q16 How do you calculate a Z-score, and what does it represent?

Ans- Z = (X - mean) / standard deviation. It measures how many standard deviations an element is from the mean.


Q17 What are point estimates and interval estimates in statistics?

- Point Estimate: Single value estimate of a parameter.
- Interval Estimate: Range of values believed to contain the parameter.


Q18 What is the significance of confidence intervals in statistical analysis?

Ans:- Indicates the reliability of an estimate, showing the range within which the parameter lies with a given probability.


Q19 What is the relationship between a Z-score and a confidence interval?

Ans- Z-scores determine the boundaries of a confidence interval for a normal distribution.


Q20 How are Z-scores used to compare different distributions?

Ans- They standardize values from different distributions for comparison.


Q21 What are the assumptions for applying the Central Limit Theorem?

- Independent sampling.
- Sufficiently large sample size.
- Finite variance.


Q22 What is the concept of expected value in a probability distribution?

Ans- The average value expected if the experiment is repeated many times.


Q23 How does a probability distribution relate to the expected outcome of a random variable?

Ans- The probability distribution defines how likely each outcome is, influencing the expected value.

```python
# Generate a random variable and display

import random
x = random.randint(1, 10)
print(x)
```

```python
#Discrete uniform distribution and plot

import numpy as np
import matplotlib.pyplot as plt

data = np.random.randint(1, 7, size=1000)
plt.hist(data, bins=np.arange(1,8)-0.5, density=True, rwidth=0.8)
plt.title('Discrete Uniform Distribution')
plt.show()
```

```python
#Bernoulli distribution PDF

def bernoulli_pdf(p, x):
    return p if x == 1 else 1 - p

print(bernoulli_pdf(0.7, 1))
```

```python
#Simulate binomial distribution (n=10, p=0.5)

from scipy.stats import binom
import matplotlib.pyplot as plt

n, p = 10, 0.5
x = np.arange(0, n+1)
plt.bar(x, binom.pmf(x, n, p))
plt.title('Binomial Distribution PMF')
plt.show()
```

```python
#Create and visualize Poisson distribution

from scipy.stats import poisson

data = poisson.rvs(mu=3, size=1000)
plt.hist(data, bins=15, density=True)
plt.title('Poisson Distribution')
plt.show()
```

```python
#Calculate and plot CDF of discrete uniform distribution

values = np.arange(1, 7)
cdf = np.cumsum([1/6]*6)
plt.step(values, cdf)
plt.title('CDF of Discrete Uniform Distribution')
plt.show()
```

```python
#Generate continuous uniform distribution and plot

samples = np.random.uniform(0, 1, 1000)
plt.hist(samples, bins=20)
plt.title('Continuous Uniform Distribution')
plt.show()
```

```python
# Simulate normal distribution data and plot

samples = np.random.normal(0, 1, 1000)
plt.hist(samples, bins=30)
plt.title('Normal Distribution')
plt.show()
```

```python
#Calculate and plot Z-scores

from scipy.stats import zscore

data = np.random.normal(5, 2, 100)
z_scores = zscore(data)
plt.plot(z_scores)
plt.title('Z-scores')
plt.show()
```

```python
#Implement Central Limit Theorem

means = []
for _ in range(1000):
    sample = np.random.exponential(scale=2, size=30)
    means.append(np.mean(sample))

plt.hist(means, bins=30)
plt.title('Central Limit Theorem Simulation')
plt.show()
```

```python
#Simulate multiple samples from a normal distribution and verify the Central Limit Theorem

import numpy as np
import matplotlib.pyplot as plt

sample_means = []
for _ in range(1000):
    sample = np.random.normal(loc=50, scale=10, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, edgecolor='black')
plt.title('Verification of Central Limit Theorem')
plt.show()
```

```python
# Calculate and plot the standard normal distribution

from scipy.stats import norm

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.title('Standard Normal Distribution')
plt.show()
```

```python
# Generate random variables and calculate probabilities using binomial distribution

n, p = 20, 0.4
x = np.arange(0, n+1)
probabilities = binom.pmf(x, n, p)
print(probabilities)
```

```python
#Calculate Z-score for a given data point and compare to standard normal distribution

x_value = 72
population_mean = 70
population_std = 5

z_score = (x_value - population_mean) / population_std
print(f"Z-score: {z_score}")
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset

sample_mean = 72
population_mean = 70
std_dev = 5
n = 30

z = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
print(f"Test Z-score: {z}")
```

```python
#Create a confidence interval for a dataset and interpret the result

import scipy.stats as stats

data = np.random.normal(50, 10, 100)
confidence = 0.95

mean = np.mean(data)
sem = stats.sem(data)
interval = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)
print(f"Confidence Interval: {interval}")
```

```python
#Calculate and interpret confidence interval for normal distribution mean

data = np.random.normal(0, 1, 100)
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)
print(f"95% Confidence Interval: {ci}")
```

```python
#Calculate and visualize PDF of a normal distribution

x = np.linspace(-3, 3, 100)
pdf = norm.pdf(x)
plt.plot(x, pdf)
plt.title('PDF of Normal Distribution')
plt.show()
```

```python
#Calculate and interpret the CDF of a Poisson distribution

from scipy.stats import poisson

cdf = poisson.cdf(np.arange(10), mu=3)
plt.step(np.arange(10), cdf)
plt.title('CDF of Poisson Distribution')
plt.show()
```

```python
# Simulate a random variable using continuous uniform distribution and calculate expected value

data = np.random.uniform(0, 10, 1000)
expected_value = np.mean(data)
print(f"Expected Value: {expected_value}")
```

```python
#Compare standard deviations of two datasets and visualize

data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(0, 2, 100)

print(f"Std dev dataset 1: {np.std(data1)}")
print(f"Std dev dataset 2: {np.std(data2)}")

plt.hist(data1, alpha=0.5, label='Std 1')
plt.hist(data2, alpha=0.5, label='Std 2')
plt.legend()
plt.title('Comparison of Standard Deviations')
plt.show()
```

```python
#Calculate range and interquartile range (IQR) of a normal distribution dataset

q75, q25 = np.percentile(data, [75 ,25])
iqr = q75 - q25
range_val = np.ptp(data)

print(f"Range: {range_val}, IQR: {iqr}")
```

```python
#Implement Z-score normalization and visualize

normalized_data = (data - np.mean(data)) / np.std(data)
plt.hist(normalized_data, bins=30)
plt.title('Z-score Normalized Data')
plt.show()
```

```python
#Calculate skewness and kurtosis of a dataset

from scipy.stats import skew, kurtosis

print(f"Skewness: {skew(data)}")
print(f"Kurtosis: {kurtosis(data)}")
```
